```python
import time

import numpy as np
import scipy.linalg as spla
import matplotlib.pyplot as plt
import torch
import GPy
```

```python
npzfile = np.load("datasets/process1a.npz")
T = npzfile['T'][:100]
Z = npzfile['Z'][:100]
plt.plot(T,Z)
T = T.astype(float)
Z = Z.astype(float)
#Most basic testing for GPs

def split_data(T,Z,p=0.5):
    ind_list = np.arange(100,dtype=int)
    inds_training = np.random.choice(ind_list,size=int(100*(1-p)),
                                     replace=False)
                                    
    inds_test = [i for i in ind_list if i not in inds_training]
    xtrain = T[inds_training]
    ytrain = Z[inds_training]
    xtest = T[inds_test]
    ytest = T[inds_test]
    return xtrain,ytrain,xtest,ytest
    
xtrain,ytrain,xtest,ytest = split_data(T,Z)
plt.plot(xtrain,ytrain,'go')
plt.show()
```

```python
def gpy_log_likelihood(xdata,ydata,h2,l,sigma2):
    xdata = xdata.reshape(-1,1)
    ydata = ydata.reshape(-1,1)
    kern = GPy.kern.RBF(input_dim=1,lengthscale=l.item(),variance=h2.item())
    model = GPy.models.GPRegression(xdata,ydata,kern,noise_var=sigma2.item())
    return model.log_likelihood()

def gpy_params(xdata,ydata,h2,l,sigma2,optimize=True):
    xdata = xdata.reshape(-1,1)
    ydata = ydata.reshape(-1,1)
    kern = GPy.kern.RBF(input_dim=1,lengthscale=l.item(),variance=h2.item())
    model = GPy.models.GPRegression(xdata,ydata,kern,noise_var=sigma2.item())
    old_time = time.clock()
    if optimize:
        model.optimize()
    new_time = time.clock()
    print(new_time - old_time)
    print(model)
    model.plot()
    plt.plot(T,Z,'r')
#     return model.log_likelihood()
```

```python
def kernel_matrix(xdata,k):
    n = xdata.shape[0]
    X1 = torch.tensor(np.repeat(xdata,n,axis=0)).float()
    X2 = torch.tensor(np.tile(xdata,[n,1])).float()
    K = k(X1,X2)
    K = K.reshape(n,n)
    return K

                      
#     K = torch.zeros(n,n)
#     for i in range(n):
#         for j in range(n):
#             K[i,j] = k(xdata[i],xdata[j])
#     return torch.tensor([[k(xi,xj) for xi in xlist] for xj in xlist])
```

```python
LOG2PI = np.log(2*np.pi)
def negative_log_likelihood(xdata,ydata,h2_sqrt,l_sqrt,sigma2_sqrt):
    h2 = h2_sqrt**2
    l = l_sqrt**2
    sigma2 = sigma2_sqrt**2
    def rbf(x1,x2):
        return h2*torch.exp(-0.5*(x1-x2)**2/(l**2))
    y_t = torch.tensor(ydata.reshape(-1,1),dtype=torch.float)
    K = kernel_matrix(xdata.reshape(-1,1),rbf)
    K += sigma2*torch.eye(len(xdata))
#     print(K)
#     invK = torch.inverse(K)
    U = torch.potrf(K)
#     print(U.detach().numpy())
    z = torch.trtrs(y_t,U,transpose=True)[0]
#     print(z.detach().numpy(),'gptorch')
    term1 = 0.5*torch.sum(z**2)
#     term1 = 
#     term2 = 0.5*torch.log(torch.prod(torch.diag(U))**2)
    term2 = torch.sum(torch.log(torch.diag(U)))
    term3 = 0.5*len(xdata)*LOG2PI
    return term1 + term2 + term3
```
